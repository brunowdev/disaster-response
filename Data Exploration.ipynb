```python
categories_sample = 'related-1;request-0;offer-0;aid_related-0;medical_help-0;medical_products-0;search_and_rescue-0;security-0;military-0;child_alone-0;water-0;food-0;shelter-0;clothing-0;money-0;missing_people-0;refugees-0;death-0;other_aid-0;infrastructure_related-0;transport-0;buildings-0;electricity-0;tools-0;hospitals-0;shops-0;aid_centers-0;other_infrastructure-0;weather_related-0;floods-0;storm-0;fire-0;earthquake-0;cold-0;other_weather-0;direct_report-0'

number_of_categories = 36

def get_categories_name_from_text(text):
    categories = list(map(lambda cat: cat.split('-')[0], text.split(';')))
    return categories

def get_categories_value_from_text(text):
    categories = list(map(lambda cat: int(cat.split('-')[1]), text.split(';')))
    return categories

categories = get_categories_name_from_text(categories_sample)

assert len(categories) == 36
```

```python

df_full = pd.merge(df_messages, df_categories, on = 'id') # pd.concat([df_messages, df_categories], keys = ['id'], axis = 1)

df_full.head(10)


df_full[categories] = df_full.categories.apply(lambda c: pd.Series(get_categories_value_from_text(str(c)))) 

df_full.head(10)
```

```python
columns_to_keep = ['id', 'message', 'genre'] + categories 
```

```python
df_full[columns_to_keep].head(10)
```

```python
df_full['length'] = df_full.message.apply(lambda msg: len(msg.strip()))
```

```python
df_full['words_length'] = df_full.message.apply(lambda msg: len(msg.strip().split(' ')))
```

```python
from nltk.corpus import words
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

import re

english_words = words.words()
lemmatizer = WordNetLemmatizer()

additional_words = ['website']

valid_words = english_words + additional_words

def lemmatize_word(word):
    return lemmatizer.lemmatize(word)

def remove_ponctuations(text):
    return re.sub(f'[!.:]', '', text)

def have_any_word(text):
    
    lower_text = text.lower()
    split_words = lower_text.split(' ')
    
    text_words = map(remove_ponctuations, split_words)
    
    lemm_words = map(lemmatize_word, text_words)
   
    return any(w in valid_words for w in lemm_words)
```

```python
possible_bad_data = df_full[(df_full['words_length'] < 10)]
bad_data = possible_bad_data[~possible_bad_data.message.apply(have_any_word)]
```

```python
bad_data.message
```

```python
df = df_full[~df_full.id.isin(list(bad_data.id))]
print(f'Rows before cleaning: {len(df_full)}')
print(f'Rows after cleaning: {len(df)}')
```

```python
df_cat_count = df[categories].sum().reset_index(name = 'sample_count').rename(columns = {
                              'index' : 'category'
                          }).sort_values(by = [ 'sample_count' ], ascending = False)

df_cat_count.head(len(df_cat_count))
```

```python
total = df_cat_count.sample_count.sum()

df_cat_count['total'] = total
```
